```python
#Merge order_df and customer_df
new_df = customer_df.merge(order_df,how="left").drop(['CustId','Age','Deals','NumWebPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth'], axis=1)
new_df.dropna(axis = 0, inplace = True)


new_df
```

```python
new_df['Age_Group'] = pd.cut(x = customer_df['Age'], bins = [17, 24, 44, 64, 150],
                         labels = [0,1,2,3])
new_df[new_df['Age_Group'] == 0]
```

```python
new_df.info()
```

```python
new_df['Income'] = new_df['Income'].apply(np.int64)
new_df.dropna(axis=0, inplace=True)
new_df.info()
```

```python
inertia = []
k = list(range(1, 10))

# Loop through each K value, find the inertia, and store it into our list
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(new_df)
    inertia.append(km.inertia_)
    # Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

```python
def get_clusters(k, data):
    # Create a copy of the DataFrame
    data = data.copy()

    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Fit the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
# check for 4 clusters
two_clusters_df = get_clusters(4,new_df)
#two_clusters[two_clusters['class'] == 1]
two_clusters_df
```

```python
# Plot a 2D graph for K = 4
two_clusters.hvplot.scatter(x="Age_Group",y="TotalPurchase" ,by="class")
```

```python
# Plot a 3D graph for K = 2
fig = px.scatter_3d(
    two_clusters,
    x="Age_Group",
    y="TotalPurchase",
    z="Income",
    color="class",
    symbol="class",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
#standardise data with StandardScaler
scaled_cust_df = StandardScaler().fit_transform(new_df)
print(scaled_cust_df[0:5])
```

```python
# Initialize PCA model, PCA method takes an argument of n_components, here the value of 3, thus reducing the features from 7 to 3
pca = PCA(n_components=3)
```

```python
#After creating the PCA model, we apply dimensionality reduction on the scaled dataset:

cust_pca = pca.fit_transform(scaled_cust_df)
cust_pca
```

```python
#Transfer PCA data to dataFrme
df_cust_pca = pd.DataFrame(
data = cust_pca,columns=["PC 1","PC 2","PC 3"]
)
df_cust_pca.head()
```

```python
# explained_variance_ratio to learn how much information can be attributed to each principal component:
# Fetch the expalined value
pca.explained_variance_ratio_
```

```python
# Find the best value for K
inertia = []
k = list(range(1, 10))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_cust_pca)
    inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# principal components data with the K-means algorithm with a K value of 5
# Initialize the K-means model
model = KMeans(n_clusters=5, random_state=0)

# Fit the model
model.fit(df_cust_pca)

# Predict clusters
predictions = model.predict(df_cust_pca)

predictions
```

```python
# Add the predicted class columns
df_cust_pca["class"] = model.labels_
df_cust_pca.head()
```

```python
#Finally, we can plot the clusters. 



# Plot a 3D graph for K = 5
fig = px.scatter_3d(
    df_cust_pca,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    color="class",
    symbol="class",
    hover_name="class",
    
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```
